```python
#remove duplicate row

df = df.drop_duplicates()

df
```

```python
#remove space

df["Product_Name"] = df["Product_Name"].str.strip()

df["Product_Name"].tolist()
```

```python
#capitalize word

df["Product_Name"] = df["Product_Name"].str.title()

df["Product_Name"].tolist()
```

```python
#Replace nan to np.nan

df["Product_Name"] = (
    df["Product_Name"]
    .astype(str)
    .replace("NaN" , np.nan)
)

df
```

```python
df["Category"].tolist()
```

```python
#remove space and capitalize word

df["Category"] = df["Category"].str.strip()
df["Category"] = df["Category"].str.title()
df["Category"].tolist
```

```python
df["Supplier"].unique()
```

```python
df["Supplier"] = (
   df["Supplier"]
    .astype(str)
    .replace({
        "jumia":"Jumia",
        "femotech":"Femtech",
        "juma":"Jumia"
    })
)

df["Supplier"].tolist()
```

```python
df
```

```python
def clean_Quantity(value):
    if value == "five":
       return 5
    elif isinstance(value ,(int , float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else : 
          return np.nan
df["Quantity"] = df["Quantity"].apply(clean_Quantity)

df["Quantity"]
```

```python
df
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"])

df["Quantity"]
```

```python
#product that needs restocking
df[["Quantity", "Reorder_Level"]]
```

```python
#which product needs restocking

df["Restock_Status"] = df.apply(
    lambda row: "Restock" if row["Quantity"] <= row["Reorder_Level"] else "YES",
    axis=1
)

df
```

```python
#what is the average unit price per category

avg_units = df.groupby("Category")["Quantity"].mean()

avg_units
```

```python
#what supplier provides the most product

supplier_count = df["Supplier"].value_counts()

supplier_count
```

```python
#Top 5 most expensive product

top_5_expensive = df.sort_values(by="Unit price($)", ascending=False).head(5)

top_5_expensive
```
